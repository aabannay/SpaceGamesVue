```python
from transformers import pipeline

# Create the translation pipeline
pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M")

# Dictionary to map languages to language codes
language_dict = {
    "English": "eng_Latn",
    "French": "fra_Latn",
    "Spanish": "spa_Latn",
    "German": "deu_Latn",
    "Italian": "ita_Latn",
    "Portuguese": "por_Latn",
    "Hindi": "hin_Deva",
    "Arabic": "ara_Arab",
    "Chinese": "zho_Hans",
    # Add more languages as needed
}

def prompt_for_language():
    print("Available languages for translation:")
    for idx, language in enumerate(language_dict.keys(), start=1):
        print(f"{idx}. {language}")

    # Get user input for source language
    src_lang_choice = int(input("Choose the number corresponding to the source language: ")) - 1
    src_lang = list(language_dict.values())[src_lang_choice]

    # Get user input for target language
    tgt_lang_choice = int(input("Choose the number corresponding to the target language: ")) - 1
    tgt_lang = list(language_dict.values())[tgt_lang_choice]

    return src_lang, tgt_lang

def translate_text():
    # Get user input for text to translate
    text = input("Enter the text to translate: ")

    # Prompt for languages
    src_lang, tgt_lang = prompt_for_language()

    # Perform translation
    result = pipe(text, src_lang=src_lang, tgt_lang=tgt_lang)

    # Display the translated text
    print(f"Translated Text: {result[0]['translation_text']}")

# Run the translation process
translate_text()
```
